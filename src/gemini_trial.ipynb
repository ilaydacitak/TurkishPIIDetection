```python
dotenv.load_dotenv()

free_api_key = os.getenv("free_api_key")
billed_api_key = os.getenv("billed_api_key")
```

```python
genai.configure(api_key=billed_api_key) # free_api_key, billed_api_key
```

```python
model = genai.GenerativeModel('gemini-2.0-flash')  # gemini-2.0-flash, gemini-2.5-flash, gemini-3-flash-preview, gemini-3-pro-preview
```

```python
prompt = f"""
Tell me 3 baby boy names that start with Er"""

print("Prompt:", prompt)
```

```python
# Send it to Gemini model
response = model.generate_content([
    prompt
])
```
